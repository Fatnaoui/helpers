```python
import json
import os
import random

labels = ["en", "it", "fr", "es", "ar_ma", "ar_msa", "ar_ma_latin", "other_ar", "other_lg"]
splits = ["train", "validation", "test"]

random.seed(42)

for split in splits:
    all_split_lines = []

    for label in labels:
        in_path = f"../MAFT/{label}/{split}.jsonl"
        if not os.path.exists(in_path):
            # some languages may not have all splits
            print(f"no {in_path}")
            continue

        with open(in_path, "r", encoding="utf-8") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue

                obj = json.loads(line)
                text = obj.get("text")
                if not text:
                    continue

                text = " ".join(text.split())
                all_split_lines.append(f"__label__{label} {text}\n")

        # os.remove(in_path)

    random.shuffle(all_split_lines)

    out_path = f"{split}.txt"
    with open(out_path, "w", encoding="utf-8") as f:
        f.writelines(all_split_lines)

    print(f"{split}: wrote {len(all_split_lines)} lines to {out_path}")

print("Done: train.txt, validation.txt, test.txt created.")
```
