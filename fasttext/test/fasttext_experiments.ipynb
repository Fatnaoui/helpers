```python
from huggingface_hub import hf_hub_download
import os

repo_id = "Fatnaoui/MAFT"
filename = "en/train.jsonl"   # change to ar_msa/train.jsonl if you prefer

path = hf_hub_download(repo_id=repo_id, repo_type="dataset", filename=filename)
print("Downloaded:", path)
print("Bytes:", os.path.getsize(path))

# Count JSONL rows
n = 0
with open(path, "rb") as f:
    for _ in f:
        n += 1
print("Rows (lines):", n)
```

```python
import fasttext

# Train a tiny supervised model
model = fasttext.train_supervised(
    input="tiny_train.txt",
    lr=0.5,
    epoch=25,
    wordNgrams=2,
)
```

```python
# Test on the training data (just to see something)
print(model.test("tiny_train.txt"))

# Predict on new examples
print(model.predict(["this movie is great"]))
print(model.predict(["this movie is awful"]))
```
